```python
import numpy as np
import math
import cv2
from matplotlib import pyplot as plt
import matplotlib.cm as cm
import tools
import pandas as pd
import sys
import os
import shutil
import random
import time
```

```python
###PATHS

##Path of BB text files
BB_files_path = '../bounding_boxes'

##Destination paths of files for the 3 sets
training_set_path_img = '../soccer_sets/training_set/img'
training_set_path_BB = '../soccer_sets/training_set/BB'
validation_set_path_img = '../soccer_sets/validation_set/img'
validation_set_path_BB = '../soccer_sets/validation_set/BB'
test_set_path_img = '../soccer_sets/test_set/img'
test_set_path_BB = '../soccer_sets/test_set/BB'
```

```python
def split():
    image_path = '../images'
    BB_path = '../bounding_boxes'
    buffer_img_path = '../buffer_img'
    buffer_BB_path = '../buffer_BB'
    
    if not os.path.isdir(buffer_img_path):
        os.makedirs(buffer_img_path)
        
    if not os.path.isdir(buffer_BB_path):
        os.makedirs(buffer_BB_path)
        
    ## Creation dossier 'buffer_img' contenant les images de elps_soccer:
    for filename in sorted(os.listdir(image_path)):
        if filename [0:11] == 'elps_soccer':
            shutil.copyfile(image_path + '/' + filename, buffer_img_path + '/' + filename)
            
    ## Creation dossier 'buffer_BB' contenant les BB de elps_soccer:
    for filename in sorted(os.listdir(BB_path)):
        shutil.copyfile(BB_path + '/' + filename, buffer_BB_path + '/' + filename)
    
    

    
    ## mettre 30% des données dans validation:
    index = []
    for filename in sorted(os.listdir(buffer_BB_path)):
        idx = filename[12:-4]
        index.append(idx)

    split = int(len(index)*0.3)
    rnd = random.sample(index, split)

    for i in rnd:
        shutil.move(buffer_BB_path + '/elps_soccer_' + i + '.txt', validation_set_path_BB + '/elps_soccer_' + i + '.txt')
        shutil.move(buffer_img_path + '/elps_soccer_' + i + '.png', validation_set_path_img + '/elps_soccer_' + i + '.png')
    

    
    
    
    ## mettre 10% des données dans test (donc 33% de validation dans test):
    index = []
    for filename in sorted(os.listdir(validation_set_path_BB)):
        idx = filename[12:-4]
        index.append(idx)

    split = int(len(index)*0.33)
    rnd = random.sample(index, split)

    for i in rnd:
        shutil.move(validation_set_path_BB + '/elps_soccer_' + i + '.txt', test_set_path_BB + '/elps_soccer_' + i + '.txt')
        shutil.move(validation_set_path_img + '/elps_soccer_' + i + '.png', test_set_path_img + '/elps_soccer_' + i + '.png')
    
    
    

    ## mettre le reste (70% des données) dans training:
    for filename in os.listdir(buffer_BB_path):
        shutil.move(buffer_BB_path + '/' + filename, training_set_path_BB + '/' + filename)
        shutil.move(buffer_img_path + '/' + filename[:-4] + '.png', training_set_path_img + '/' + filename[:-4] + '.png')



  
```

```python
##Create directories for the 3 sets:

if not os.path.isdir(training_set_path_img):
        os.makedirs(training_set_path_img)
        
if not os.path.isdir(training_set_path_BB):
        os.makedirs(training_set_path_BB)  
        
if not os.path.isdir(validation_set_path_img):
        os.makedirs(validation_set_path_img)
        
if not os.path.isdir(validation_set_path_BB):
        os.makedirs(validation_set_path_BB)
        
if not os.path.isdir(test_set_path_img):
        os.makedirs(test_set_path_img)
        
if not os.path.isdir(test_set_path_BB):
        os.makedirs(test_set_path_BB)        
        


split()


##Supprimer les 2 buffers:
buffer_img_path = '../buffer_img'
buffer_BB_path = '../buffer_BB'
os.rmdir(buffer_img_path)
os.rmdir(buffer_BB_path)


print("Training set, validation set and test set created")
```
